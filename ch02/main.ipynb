## and_gate.py

### 簡単な実装

```python
def AND(x1, x2):
  w1, w2, theta = 0.5, 0.5, 0.7
  tmp = x1*w1 + x2*w2
  if tmp <= theta:
    return 0
  elif tmp > theta:
    return 1
```

```python
print(AND(0, 0)) # 0を出力
print(AND(1, 0)) # 0を出力
print(AND(0, 1)) # 0を出力
print(AND(1, 1)) # 1を出力
```

### 重みとバイアスによる実装

```python
import numpy as np

def AND(x1, x2):
  x = np.array([x1, x2])
  w = np.array([0.5, 0.5])
  b = -0.7
  tmp = np.sum(w*x) + b
  if tmp <= 0:
    return 0
  else:
    return 1
```

```python
print(AND(0, 0)) # 0を出力
print(AND(1, 0)) # 0を出力
print(AND(0, 1)) # 0を出力
print(AND(1, 1)) # 1を出力
```

## nand_gate.py

```python
import numpy as np

def NAND(x1, x2):
  x = np.array([x1, x2])
  w = np.array([-0.5, -0.5]) # 重みとバイアスだけがANDと違う！
  b = 0.7
  tmp = np.sum(w*x) + b
  if tmp <= 0:
    return 0
  else:
    return 1
```

```python
print(NAND(0, 0)) # 1を出力
print(NAND(1, 0)) # 1を出力
print(NAND(0, 1)) # 1を出力
print(NAND(1, 1)) # 0を出力
```

## or_gate.py

```python
import numpy as np

def OR(x1, x2):
  x = np.array([x1, x2])
  w = np.array([0.5, 0.5]) # 重みとバイアスだけがANDと違う！
  b = -0.2
  tmp = np.sum(w*x) + b
  if tmp <= 0:
    return 0
  else:
    return 1
```

```python
print(OR(0, 0)) # 0を出力
print(OR(1, 0)) # 1を出力
print(OR(0, 1)) # 1を出力
print(OR(1, 1)) # 1を出力
```

## xor_gate.py

```python
def XOR(x1, x2):
  s1 = NAND(x1, x2)
  s2 = OR(x1, x2)
  y = AND(s1, s2)
  return y
```

```python
print(XOR(0, 0)) # 0を出力
print(XOR(1, 0)) # 1を出力
print(XOR(0, 1)) # 1を出力
print(XOR(1, 1)) # 0を出力
```
